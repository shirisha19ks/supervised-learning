```python
for c in dfhr.select_dtypes(include=['int64','float64']).columns:
    
    Q1 = dfhr[c].quantile(0.25)
    Q3 = dfhr[c].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    dfhr[c] = dfhr[c].apply(
        lambda x: lower if x < lower else upper if x > upper else x
    )
```
